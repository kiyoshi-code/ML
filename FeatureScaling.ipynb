```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"]=pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]),columns=["x"])
dataset.head(3)
```

```python
sns.displot(dataset["CoapplicantIncome"], kde=True) #kde for smooth curve
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
dataset["CoapplicantIncome_min"]=ms.transform(dataset[["CoapplicantIncome"]])
dataset.head(3)
             
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("before")
#sns.displot(dataset["CoapplicantIncome"])
sns.histplot(dataset["CoapplicantIncome"], kde=True)
#sns.histplot() draws on the current plt.subplot()
#sns.displot() creates a new figure, so it's not meant for use with plt.subplot()

plt.subplot(1,2,2)
plt.title("after")
sns.histplot(dataset["CoapplicantIncome_min"], kde=True)
plt.tight_layout()
plt.show()
```
