```python
#logistic-use when linearly seperable inputs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv(r"C:\Users\rasgu\OneDrive\Documents\VS Code\ml\datasets\Social_Network_Ads.csv")
dataset.drop(columns=["EstimatedSalary"], inplace=True)
dataset.head(3)
```

```python
#logistic applicable?:
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age", y="Purchased", data=dataset)
plt.show()
```

```python
x=dataset[["Age"]]
y=dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[40]])
#op=array([0])=no
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age", y="Purchased", data=dataset)
sns.lineplot(x="Age", y=lr.predict(x), data=dataset, color='red')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset1=pd.read_csv(r"C:\Users\rasgu\OneDrive\Documents\VS Code\ml\datasets\placement-dataset.csv")
dataset1.head(3)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa", y="score", data=dataset1, hue="placement")
#plt.legend(loc=1)
plt.show()
```

```python
#check linearly seperable or not:
#seperating as dependent and independent:

x=dataset1.iloc[:,:-1]
y=dataset1["placement"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[6.8, 123]])
#now graph has been linearly seperated
```
