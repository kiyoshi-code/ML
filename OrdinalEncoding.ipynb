```python
import pandas as pd
df=pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
df.head(3)
```

```python
order_data=[["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=order_data)
oe.fit_transform(df[["Size"]])
```

```python
df["size_en"]=oe.transform(df[["Size"]])
df
```

```python
#map function
ord_data={"s":4,"m":5,"l":6,"xl":7} #dictionary, can assign nos to data acc to our choice
df["Size"].map(ord_data)
```

```python
df["Size_en_map"]=df["Size"].map(ord_data)
df
```
