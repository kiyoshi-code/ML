```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv(r"C:\Users\rasgu\OneDrive\Documents\VS Code\ml\datasets\polynomial_classification.csv")
dataset.head(3)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="data1", y="data2", data=dataset, hue="output")
plt.show()
```

#simple logistic:

```python
x=dataset.iloc[:,:-1]
y=dataset["output"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=lr)
plt.show()
#not properly classifed
```

#with polynomial feature

```python
x=dataset.iloc[:,:-1]
y=dataset["output"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=2)
pf.fit(x)
x=pd.DataFrame(pf.transform(x))
x
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test, y_test)*100
```
