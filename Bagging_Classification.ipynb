```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
```

```python
x,y=make_moons(n_samples=1000, noise=0.2)
df={"x1":x[:,0],"x2":x[:,1], "y":y}
dataset=pd.DataFrame(df)
dataset.head(5)
```

```python
sns.scatterplot(x="x1", y="x2", hue="y", data=dataset)
plt.show()
```

```python
x_a=dataset.iloc[:,:-1]
y_a=dataset["y"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x_a,y_a,random_state=42, test_size=0.2)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
bg=BaggingClassifier(estimator=SVC(), n_estimators=100)
bg.fit(x_train, y_train)
```

```python
bg.score(x_train, y_train)*100, bg.score(x_test, y_test)*100
```

```python
sv=SVC()
sv.fit(x_train , y_train)
sv.score(x_train, y_train)*100, sv.score(x_test, y_test)*100
```

Rndom Forest

```python
rf=RandomForestClassifier(n_estimators=30)#default=decision tree
rf.fit(x_train, y_train)
rf.score(x_train,y_train)*100, rf.score(x_test, y_test)*100
```
