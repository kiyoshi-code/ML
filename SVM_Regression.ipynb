```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv(r"C:\Users\rasgu\OneDrive\Documents\VS Code\ml\datasets\placement.csv")
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="cgpa", y="package", data=dataset)
plt.show()
```

```python
x=dataset[["cgpa"]]
y=dataset["package"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
from sklearn.svm import SVR
sv=SVR(kernel="linear")
#linear, poly, rbf, sigmoid
sv.fit(x_train, y_train)
```

```python
sv.score(x_test, y_test)*100
```

```python
sv.score(x_train, y_train)*100
```

```python
sns.scatterplot(x="cgpa", y="package", data=dataset)
plt.plot(dataset["cgpa"], sv.predict(x), color="black")
plt.show()
```

```python
sv1=SVR(kernel="poly", degree=3)
sv1.fit(x_train, y_train)
```

```python
sv1.score(x_test, y_test)*100
```

```python
sv1.score(x_train, y_train)*100
```

```python
sns.scatterplot(x="cgpa", y="package", data=dataset)
plt.plot(dataset["cgpa"], sv1.predict(x), color="black")
plt.show()
plt.scatter(dataset["cgpa"], sv1.predict(x), color="black")
plt.show()
```
