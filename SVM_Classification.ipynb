```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv(r"C:\Users\rasgu\OneDrive\Documents\VS Code\ml\datasets\polynomial_classification.csv")
dataset.head(5)
```

```python
dataset.isnull().sum()
#no encoding or scalling required
```

```python
#check linearly or non-linearly seperable data
plt.figure(figsize=(4,3))
sns.scatterplot(x="data1", y="data2", data=dataset, hue="output")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["output"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,random_state=42, test_size=0.2)
```

```python
from sklearn.svm import SVC
sv=SVC(kernel="rbf")
#kernel=linear, poly, rbf, precomputed, sigmoid
sv.fit(x_train, y_train)
#for precomputed, most data should be in 0-1
```

```python
sv.score(x_test, y_test)*100
```

```python
sv.score(x_train, y_train)*100
#changing random state also helps
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=sv)
plt.show()
```
