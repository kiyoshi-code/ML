```python
import pandas as pd
```

```python
dataset=pd.read_csv(r"C:\Users\rasgu\OneDrive\Documents\VS Code\ml\datasets\Social_Network_Ads.csv")
dataset.head(3)
```

```python
dataset["Purchased"].value_counts()
#to show imbalancing in data
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,random_state=42, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
#to check if it has become biased or not
lr.predict([[19, 19000]])
```

```python
lr.predict([[45,26000]]) #should give 1 but gives 0 because of imbalancing-data has become biased-so balance data
```

```python
from imblearn.under_sampling import RandomUnderSampler
#takes 2 input data and returns 4
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Purchased"]
```

```python
ru=RandomUnderSampler()
ru_x, ru_y=ru.fit_resample(x,y)
ru_x, ru_y
```

```python
ru_y.value_counts()
#this gives new x and y that are ru_x and ru_y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train1, x_test1, y_train1, y_test1=train_test_split(ru_x, ru_y, random_state=42, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr1=LogisticRegression()
lr1.fit(x_train1, y_train1)
```

```python
lr1.score(x_test1, y_test1)*100
```

```python
lr1.predict([[46, 28000]])
```

```python
from imblearn.over_sampling import RandomOverSampler
ro=RandomOverSampler()
ro_x, ro_y=ro.fit_resample(x,y)
```

```python
ro_y.value_counts()
```

```python
x_train2, x_test2, y_train2, y_test2=train_test_split(ro_x,ro_y, random_state=42, test_size=0.2)
```

```python
lr2=LogisticRegression()
lr2.fit(x_train2, y_train2)
```

```python
lr2.score(x_test2,y_test2)*100
```

```python
lr2.predict([[46,28000]])
```
