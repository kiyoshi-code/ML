```python
import pandas as pd
```

```python
dataset=pd.read_csv(r"C:\Users\rasgu\OneDrive\Documents\VS Code\ml\datasets\placement.csv")
dataset.head(5)
```

```python
x=dataset.iloc[:,:-1]
y=dataset["package"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,random_state=42, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train, y_train)*100, lr.score(x_test, y_test)*100
```

```python
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
dt.score(x_train, y_train)*100, dt.score(x_test, y_test)*100
#but here overfitting is there
```

```python
sv=SVR()
sv.fit(x_train,y_train)
sv.score(x_train, y_train)*100, sv.score(x_test, y_test)*100
```

```python
from sklearn.ensemble import VotingRegressor
li=[("lr1",LinearRegression()),("dt1",DecisionTreeRegressor()),("sv1",SVR())]

vc=VotingRegressor(li)
vc.fit(x_train, y_train)
```

```python
vc.score(x_train, y_train)*100, vc.score(x_test, y_test)*100
#overfitting is still there, so svr has dominated; tuning might inc accuracy
```

```python
#cross checking:
df={"lr":lr.predict(x_test),"dt":dt.predict(x_test), "svm":sv.predict(x_test), "vc":vc.predict(x_test)}
pd.DataFrame(df).head(5)
```
