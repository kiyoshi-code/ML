```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset=pd.read_csv(r"C:\Users\rasgu\OneDrive\Documents\VS Code\ml\datasets\placement-dataset.csv")
dataset.head(5)
```

```python
#check null values
dataset.isnull().sum()
#scalling not required in this data
```

```python
#graph:
plt.figure(figsize=(4,3))
sns.scatterplot(x="cgpa", y="score", data=dataset, hue="placement")
plt.show()
#graph is linearly seperale
```

```python
x=dataset.iloc[:,:-1]
y=dataset["placement"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,random_state=42, test_size=0.2)
```

```python
#kd plot=distribution plot
sns.kdeplot(data=dataset["cgpa"])
plt.show()
```

```python
sns.kdeplot(data=dataset["score"])
plt.show()
```

```python
#based on data apply type of naive bayes
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
gnb=GaussianNB()
gnb.fit(x_train, y_train)
```

```python
gnb.score(x_test, y_test)*100, gnb.score(x_train, y_train)*100
```

```python
gnb.predict([[6.8,123]])
```

```python
mnb=MultinomialNB()
mnb.fit(x_train, y_train)
```

```python
mnb.score(x_test, y_test)*100, mnb.score(x_train, y_train)*100
```

```python
bnb=BernoulliNB()
bnb.fit(x_train, y_train)
```

```python
bnb.score(x_test, y_test)*100, bnb.score(x_train, y_train)*100
```

```python
#deciding with help of decision region:
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=gnb)
plt.show()
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=mnb)
plt.show()
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=bnb)
plt.show()
```
